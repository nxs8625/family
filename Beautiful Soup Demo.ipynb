```python
from bs4 import BeautifulSoup
import urllib.request
```

```python
f = open('outfile538.txt','w')
errorFile = open('errorfile538.txt','w')
```

```python
soup = BeautifulSoup(urllib.request.urlopen('https://www.fantasypros.com/nba/player-rater.php').read(),'html.parser')
#print(soup)
```

```python
tableStats = soup.find("div", class_= "mobile-table") # table-bordered table-striped table-hover player-table tablesorter")

#print(tableStats)
```

```python
for row in tableStats.findAll('tr'):
    col = row.findAll('td')
    
    try:
        name = col[1].a.string.strip()
        fgpct = col[2].string.strip()
        # and so on and so on
        overall = col[10].string.strip()
        
        print(name,fgpct, overall)
        f.write(name + '\n')
        
    except Exception as e:
        errorFile.write('error' + str(e))
        pass

f.close
errorFile.close
```

```python
import pandas as pd
header = soup.find('thead')
columns = [col.get_text() for col in header.find_all('th')]
finaldf = pd.DataFrame()

```

```python
for row in tableStats.findAll('tr'):
    stats = [stat.get_text() for stat in row.findAll('td')]
    tempdf = pd.DataFrame(stats).transpose()
    finaldf = pd.concat([finaldf, tempdf], ignore_index=True)
finaldf.columns=columns
print(finaldf)
```
